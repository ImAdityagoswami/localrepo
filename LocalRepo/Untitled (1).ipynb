```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

Q1. Top 10 batsmen by total runs

```python
top_runs = df.groupby('PlayerName')['Runs'].sum() \
             .sort_values(ascending=False).head(10)

top_runs

top_runs.plot(kind='bar')

plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Top 10 Batsmen by Total Runs")

plt.show()
```

Q2. Kis player ka strike rate sabse best hai?

```python
df['StrikeRate'] = pd.to_numeric(df['StrikeRate'], errors='coerce')

best_sr = df.groupby('PlayerName')['StrikeRate'].mean()\
            .sort_values(ascending=False).head(10)

best_sr
```

Q3. Sabse consistent batsman kaun hai?

```python
consistent = df.groupby('PlayerName')['Runs'].mean()\
    .sort_values(ascending=False).head(10)
consistent
```

Q4. Team-wise total runs comparison


```python
comparison = df.groupby('Team')['Runs'].sum()\
                .sort_values(ascending=False)
print(comparison)

comparison.plot(kind='bar')

plt.xlabel('Teams')
plt.ylabel('Total Runs')
plt.title('Team-wise total runs comparison')

plt.show()
```

Q5. Kis team ne sabse jyada sixes maare?


```python
team_sixes = df.groupby('Team')['Sixes'].sum()\
                .sort_values(ascending=False)
print(team_sixes)

team_sixes.plot(kind='bar')

plt.xlabel('Teams')
plt.ylabel('Total_Sixes')
plt.title('Team-Wise-Sixes')
plt.show()
```

Q6. Kis player ne sabse jyada boundaries lagayi?


```python
df['Boundaries'] = df['Sixes'] + df['Fours']
boundaries = df.groupby('PlayerName')['Boundaries'].sum()\
                    .sort_values(ascending=False).head(10)
print(boundaries)
boundaries.plot(kind='bar')

plt.xlabel('Players')
plt.ylabel('Boundaries')
plt.title('Top-Players-by-Boundaries')

plt.show()
```

Q7. Balls vs Runs relationship analysis


```python
sns.scatterplot(x=df['Balls'],y=df['Runs'])
plt.xlabel('Balls')
plt.ylabel('Runs')
plt.title('Balls vs Runs relationship')

plt.show()
```

Q10. Spin vs Pace ke against best performer kaun hai?


```python
spin_against = df.groupby('PlayerName')['AgainstSpin'].sum()\
                    .sort_values(ascending=False).head(10)

fast_against = df.groupby('PlayerName')['AgainstFast'].sum()\
                .sort_values(ascending=False).head(10)
print(spin_against)
print(fast_against)
```

Q11. Dot balls ka impact on strike rate


```python
sns.scatterplot(x=df['DotBalls'],y=df['StrikeRate'])
plt.xlabel('DotBalls')
plt.ylabel('StrikeRate')
plt.title('DotBalls vs StrikeRate')

plt.show()
```

Q12. Highest average of player


```python
high_avg = df.groupby('PlayerName')['Runs'].mean()\
                .sort_values(ascending=False).head(10)
print(high_avg)

high_avg.plot(kind='bar')

plt.xlabel('PlayerName')
plt.ylabel('Runs')
plt.title('Highest-Average-Player')

plt.show()
```

Q14. Team-wise batting strength comparison


```python
mvp = df.groupby('PlayerName')[['Runs','StrikeRate']].mean()

mvp['impact'] = mvp['Runs'] * mvp['StrikeRate']

mvp = mvp.sort_values('impact',ascending=False).head(10)

print(mvp)
```

Q15. Aggressive vs anchor batsmen analysis


```python
aggressive = df[df['StrikeRate'] > 150]

anchor = df[df['StrikeRate'] < 120]

print('Aggressive batsmen')

print(aggressive[['PlayerName','StrikeRate']].head(10))

print('Anchor batsmen')

print(anchor[['PlayerName','StrikeRate']].head(10))
```

Q16. Top finishers ka analysis


```python
top_finishers = df.groupby('PlayerName')['Runs'].sum()\
                    .sort_values(ascending=False).head(10)
print(top_finishers)

top_finishers.plot(kind='bar')

plt.xlabel('PlayerName')
plt.ylabel('Runs')
plt.title('Top finishers ka analysis')

plt.show()
```

Q17. Identify match-winning innings

Q18. Runs distribution histogram


```python
sns.histplot(df['Runs'],bins=20)
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.title('Runs distribution histogram')

plt.show()
```

Q19. Team-wise batting heatmap


```python
heatmap_data = df.groupby('Team')[['Runs','StrikeRate',
                                   'Sixes',
                                  'Fours']].mean()
sns.heatmap(heatmap_data,annot=True)

plt.title('Team-wise batting heatmap')

plt.show()
```

Q20. Player performance trend visualization

```python
player_performance = df.groupby('PlayerName')['Runs'].sum()\
                        .sort_values(ascending=False).head(10)
player_performance.plot(kind='line',marker='o')

plt.xlabel('Players')
plt.ylabel('Runs')
plt.title('20. Player performance trend visualization')
plt.show()
```
